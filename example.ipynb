```python
import matplotlib.pyplot as plt
import numpy as np
import time
from WKMeansClustering import WKMeans, KMeans
```

```python
# importer les donn√©es 
from sklearn.datasets import make_blobs

n_samples = 1000

random_state = 170 

X, y = make_blobs(n_samples=n_samples, random_state=random_state, shuffle=False)

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
y_knn = KMeans(n_cluters=3).fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_knn)
plt.show()
```

```python
start = time.time()

w = np.ones(n_samples) /n_samples

wkmean = WKMeans(n_cluters=3, w=w)

y_wknn = wkmean.fit_predict(X)

cluster_cunters = wkmean.cluster_centers_

print(f"temps d'execution: {time.time() - start}s")


plt.scatter(X[:, 0], X[:, 1], c=y_wknn)
plt.scatter(cluster_cunters[:, 0], cluster_cunters[:, 1], c='red', s=100)
plt.show()
```
